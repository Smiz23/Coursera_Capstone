# Question 1

### 1. Retrieving dataset using Pandas 

```python
import pandas as pd

data = pd.read_html(io='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',match='Postcode')
```

### 2. Data Cleaning

```python
df = pd.DataFrame(data[0])
df.columns = ['PostalCode','Borough','Neighborhood']
df = df.loc[df['Borough']!= 'Not assigned']
df = df.reset_index()
df.drop(['index'],axis=1)
```

### 3. Final Dataframe

```python
df1 = df.groupby(['PostalCode','Borough'],as_index=False,sort=False).agg(','.join)
df1
```

### 4. Shape of the dataframe

```python
df1.shape
```

```python
df2 = df1.sort_values(by=['PostalCode'],ascending = True,axis=0)
df2 = df2.reset_index()
df2 = df2.drop(['index'],axis=1)
df2
```

* Retrieve top venues for each neighborhood

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
import numpy as np
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
df7 = pd.DataFrame(columns=columns)
df7['Neighborhood'] = df6['Neighborhood']

for i in np.arange(df6.shape[0]):
    df7.iloc[i, 1:] = return_most_common_venues(df6.iloc[i, :], num_top_venues)

df7.head()
```

* Clustering using KMeans method

```python
df9 = df6.drop(['Neighborhood'],axis=1)
```

```python
import sklearn
from sklearn.cluster import KMeans
df8 = KMeans(init='k-means++',n_clusters=5).fit(df9)
df8.labels_[0:10]
```
