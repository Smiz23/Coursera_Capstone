## Data Wrangling

The above table has few improper formats.So, the unnecessary columns like venue referral id, country etc are removed. Adding on this, the categories column is an inbuilt array with the venue id and the category as the elements. This column has to be segregated , hence, the below function is used to generate a new column for Category. 

```python
def getcategory(i):
    
                    try:
                        c_list = i['categories']
                    except:
                        c_list = i['venue.categories']
        
                    if len(c_list) == 0:
                                    return None
                    else:
                        return c_list[0]['name']
```

A new list called School_list is generated with the specific columns and the category alone for further analysis

```python

col = ['id','name','location.distance','location.address','location.lat','location.lng']
school_list =venue_list.loc[:, col]

school_list['Category'] = venue_list.apply(getcategory, axis=1)

school_list
```

The above list has few 'NAN' values in the Address location. Hence, those rows are removed and the index is reset as shown below.

```python
school_list = school_list.dropna(subset=['location.address'],axis=0)
school_list = school_list.dropna(subset=['location.address'],axis=0)
school_list=school_list.reset_index()

school_list = school_list.drop(['index'],axis=1)
school_list
```

## Data Modelling - K-Means Clustering Model

Map data is basically unlabelled data, which means we cannot have a specific target variable or an independent variable. We are looking only for schools/universities that is nearby for the user. Hence, Exploratory Data analysis is not required. For unlabelled data, clustering techniques are utilised. The objective of the project is to split data in the form of different clusters based on the category obtained in the previously displayed table.

### Clustering Model

Initially, the data had to be more categorized to imply the machine learning algorithms. Basically, looking at the table, I displayed earlier, we may not require latitude, longitude, name of the school for our modelling. Only the category was required. Hence,One hot encoding technique is utilised to convert the categorical variables into numerical variables for data modelling. Upon executing the script, below table is generated.


```python

clust_list = pd.get_dummies(school_list[['Category']], prefix="", prefix_sep="")


clust_list['Address'] = school_list['location.address'] 

fixed_columns = [clust_list.columns[-1]] + list(clust_list.columns[:-1])
clust_list = clust_list[fixed_columns]

clust_list
```

### K-Means Algorithm

For the above data, the most suitable clustering technique is K-Means algorithm. This algorithm is used since we required the Euclidean distance between each of the locations. The number of clusters to be generated is set as 4. Upon applying the algorithm to the dataset, the records are split as cluster types displayed below.

```python
kclusters = 4

cluster =clust_list.drop('Address', 1)


kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(cluster)


kmeans.labels_[0:10] 
```

A final list containing all the required data is set for Model evaluation.

```python
school_list.insert(0, 'Cluster Labels', kmeans.labels_)

school_list
```

## Model Evaluation

Using the Home location column, I searched the location address in the school list we generated earlier. Since, the find function returns only '-1' or any positive value, I used the index value of those selected records and stored them in a dataframe as displayed below

```python
u_list = pd.DataFrame(columns=['location.address'])
for i,j in zip(user_list['Home Location'],user_list['School Rating']):

                                                                    s = pd.DataFrame(school_list['location.address'].str.find(i))
                                                                    s1 = s.loc[s['location.address'] != -1]
                                                                    u_list = u_list.append(s1)
                                                                     
u_list = u_list.sort_index()
u_list
```

### User Output

The table generated above is used to display the Map containing the select schools according to User preference.


```python
user_map = folium.Map(location=[latitude, longitude], zoom_start=14)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster,category in zip(final_list['location.lat'], final_list['location.lng'], final_list['name'],final_list['Cluster Labels'],final_list['Category']):
    label = folium.Popup(str(poi) + ',' + str(category), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=10,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(user_map)
       
user_map
```
